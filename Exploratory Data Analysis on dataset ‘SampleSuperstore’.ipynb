```python
# Importing all the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings as warn
warn.filterwarnings("ignore")
```

```python
# Reading data from the link
link="https://raw.githubusercontent.com/Gourish14/Exploratory-Data-Analysis-on-dataset-SampleSuperstore-/main/SampleSuperstore.csv"
df = pd.read_csv(link)
```

```python
# Reading the dataset
df.head()
```

```python
# Data Preprocessing
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Checking for duplicate values
df.duplicated().sum()
```

```python
# Dropping the duplicates
df.drop_duplicates()
df.head()
```

```python
# Removing the unnecessary columns i.e. postal code
df = df.drop(['Postal Code'],axis=1)
```

```python
# Reading the dataset after removing postal code
df.head()
```

```python
# Visualizing the whole dataset using pair plot
import seaborn as sns 
sns.pairplot(df)
```

```python
# Finding the pairwise correlations b/w columns and visualisation using heatmaps
df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
# Displaying in the form of Pie-Chart
textprops = {"fontsize":20}
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```python
# Displaying in the form of Plot Bar
plt.figure(figsize= (10,16))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
# Computing top categories in terms of sales from first 100 observations
top_category_s = df.groupby("Category").Sales.sum().nlargest(n=100)
# Computing top categories in terms of profit from first 100 observations
top_category_p = df.groupby("Category").Profit.sum().nlargest(n=100)

# Plotting to see these visually
plt.style.use('seaborn')
top_category_s.plot(kind = 'bar',figsize = (10,5),fontsize = 20)
top_category_p.plot(kind = 'bar',figsize = (10,5),fontsize = 20,color='green')
plt.xlabel('Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Category Sales vs Profit",fontsize = 20)
plt.show()
```

```python
# Computing top sub-categories in terms of sales from first 100 observations
top_subcategory_s = df.groupby("Sub-Category").Sales.sum().nlargest(n = 100)
# Computing top sub-categories in terms of profit from first 100 observations
top_subcategory_p = df.groupby("Sub-Category").Profit.sum().nlargest(n = 100)

# Plotting to see these visually
plt.style.use('seaborn')
top_subcategory_s.plot(kind = 'bar',figsize = (10,5),fontsize = 20)
top_subcategory_p.plot(kind = 'bar',figsize = (10,5),fontsize = 20, color = 'green')
plt.xlabel('Sub-Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Sub-Category Sales vs Profit",fontsize = 20)
plt.show()
```

```python
# More detailed view
plt.figure(figsize=(14,12))
statewise = df.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh()
```

```python
# Visualising the discount
plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'], df['Profit'], data=df)
```

```python
# Visualizing the Sales vs Profits in different Regions
plt.figure(figsize = (6,6))
plt.title('Region')
plt.pie(df['Region'].value_counts(), labels=df['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
# Computing top states in terms of sales from first 10 observations
top_states_s = df.groupby("State").Sales.sum().nlargest(n=10)

# Computing top states in terms of profit from first 10 observations
top_states_p = df.groupby("State").Profit.sum().nlargest(n = 10)

plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 20)
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 20, color = 'green')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 20)
plt.show()
```

```python
# Checking the interdependency of Sales,Profits and Discounts
plt.style.use('seaborn')
df.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 20, colormap = "viridis")
plt.ylabel('Total Profits',fontsize = 15)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 15)
plt.show()
```
